```python
dummy_data1 = {'Age':['11','22','35','23','67'],
               'City' : ['Mumbai','Pune','Chennai','Banglore','Delhi'],
               'Aadhar':['123','321','213','132','412']}
df1 = pd.DataFrame(dummy_data1,columns = ['Age','City','Aadhar'])
df1
```

```python
dummy_data2 = {
    'Age':['21','18','53','14','10'],
    'City': [ 'Pimpri','Nandurbar','Shahada','Pune','Chennai'],
    'Aadhar':['890','897','789','987','459']
}

df2 = pd.DataFrame(dummy_data2,columns = ['Age','City','Aadhar'])
df2
```

```python
df_aadhar = pd.concat([df1,df2],ignore_index= True)
df_aadhar
```

```python
dummy_data3 = {
    'Aadhar' : ['123','321','213','132','412','897','789','987','459','12345'],
    'Gender' : ['Male','Female','Male','Female','Male','Female','Female','Male','Male','Female']
}
df3 = pd.DataFrame(dummy_data3,columns = ['Aadhar','Gender'])
df3
```

```python
df_merge = pd.merge(df_aadhar,df3,on = 'Aadhar')
df_merge
```

```python
dummy_datapune = {
    'Age':['11','12','35'],
    'City' : ['Mumbai','Pune','Chennai'],
    'Aadhar' : ['123','234','456']
}
df_pune = pd.DataFrame(dummy_datapune,columns=['Age','City','Aadhar'])
df_pune
```

```python
dummy_data_kothrud = {
    'Aadhar_kothrud': [ '123','234','567'],
    'Gender':['Male','Female','Male']
}
df_kothrud = pd.DataFrame(dummy_data_kothrud,columns=['Aadhar_kothrud','Gender'])
df_kothrud
```

```python
df_merge_pune = pd.merge(df_pune,df_kothrud,left_on='Aadhar',right_on='Aadhar_kothrud')
df_merge_pune
```
